### naive function  

```python
import math
#simple euclidean distance calculation : manual 
def calculate_euclidean_distance_1(coords1,coords2):
    for i in range(0,len(my_coords1)):
        d1=(coords1[i][0]-coords2[i][0])** 2
        d2=(coords1[i][1]-coords2[i][1])** 2
        distance = math.sqrt(d1+d2)
        print("Euclidean distance from x to y: ",distance)
```

```python
calculate_euclidean_distance_1(my_coords1,my_coords2)
```

```python
%timeit -n 100 calculate_euclidean_distance_1(my_coords1,my_coords2)
```

### using math lib and iterators

```python
import math
#simple euclidean distance calculation : manual 
def calculate_euclidean_distance_2(coords1,coords2):
    for i in range(0,len(my_coords1)):
        distance = math.sqrt(sum([(a - b) ** 2 for a, b in zip(coords1[i],coords2[i])]))
        print("Euclidean distance from x to y: ",distance)
```

```python
calculate_euclidean_distance_2(my_coords1,my_coords2)
```

```python
%timeit -n 100 calculate_euclidean_distance_2(my_coords1,my_coords2)
```

```python
%%writefile hw_q1.py

import numpy as np
import sys


from mpi4py import MPI
comm = MPI.COMM_WORLD

size = comm.Get_size()
rank = comm.Get_rank()  

#var= np.array([0,])
   
#A process with rank r > 0 does the following: 
#    (i) receives a value say v; 
#    (ii) computes v:= v*r;  and 
#    (iii) sends v to Process (r+1); except that the last process sends v to Process 0.

var = np.zeros(1)


if rank == 0:
    print('please enter and integer between 0-100')
    for line in sys.stdin:
        var = np.array(int(line))
        if 0 <= var and var<= 100: # the condition 
            print('condition met')
            break
    comm.send(var, dest=1, tag=0)
    print('process- 0 sent:',var)
    data = comm.recv(source=size-1, tag=0)
    print("process- 0 received:",data)
    
while((rank>0)&(rank<size-1)):
    data = comm.recv(source=rank-1, tag=0)
    print("process-",rank,"received:",data)
    v = rank*data
    #print(var[0])
    comm.send(np.array(v), dest=rank+1, tag=0)
    print("process-",rank,"sent:",v)

if (rank==size-1):
    #print("receiving....")
    data2 = comm.recv(source=rank-1, tag=0)
    print("process-",rank,"received:",data2)
    v = rank*data2
    #print(var[0])
    comm.send(np.array(v), dest=0, tag=0)
    print("process-",rank,"sent to dest-0:",v)
    
```
